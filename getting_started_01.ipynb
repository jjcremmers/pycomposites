# Code

Now, the first lines of the python code can be written in a python cell.

```python
from math import sin,cos,pi,sqrt,tan,atan
from numpy import zeros
from pylab import plot, show, xlabel, ylabel
```

```python
xplot = zeros(100)
y1plot = zeros(100)
y2plot = zeros(100)
y3plot = zeros(100)

for i in range(100):
    xplot[i] = i*2*pi/100
    y1plot[i] = sin(xplot[i]) 
    y2plot[i] = cos(xplot[i])**2
    y3plot[i] = 0.25*(xplot[i]-pi)**2
      
plot( [x for x in xplot], [y for y in y1plot], 'r-' )
plot( [x for x in xplot], [y for y in y2plot], 'b' )
plot( [x for x in xplot], [y for y in y3plot], 'g' )

show()
```
